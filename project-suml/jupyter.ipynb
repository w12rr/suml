```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
from autogluon.tabular import TabularDataset, TabularPredictor
from sklearn.model_selection import train_test_split
```

```python
# using the Csv file 
df = pd.read_csv('Car_sale_ads_short.csv')  
  
# Checking the first 5 entries of dataset 
df.head() 
```

```python
df.drop(columns=['Index', 'Offer_publication_date', 'Offer_location', 'Features' ,'Origin_country', 'First_owner', 'First_registration_date'], inplace=True)
print(df.head())
```

```python
train, test = train_test_split(df, test_size=0.2)
```

```python
train_data = TabularDataset(train)
test_data = TabularDataset(test)
test_data.drop(columns=['Price'], inplace=True)
```

```python
predictor = TabularPredictor(label='Price', path='predictor', problem_type='regression').fit(train_data, presets='medium_quality')
```

```python
predictions = predictor.predict(test_data)
print(predictions)
print(predictor.leaderboard())
```
